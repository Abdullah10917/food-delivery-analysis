# Food Delivery Data Analysis Hackathon

This notebook demonstrates loading, merging, and analyzing food delivery datasets
from CSV, JSON, and SQL sources as per the given problem statement.


```python
import pandas as pd
import sqlite3
```

```python
orders = pd.read_csv("orders.csv")
orders.head()
```

```python
users = pd.read_json("users.json")
users.head()
```

```python
conn = sqlite3.connect("restaurants.db")
restaurants = pd.read_sql("SELECT * FROM restaurants", conn)
restaurants.head()
```

```python
merged = orders.merge(users, on="user_id", how="left")
final_df = merged.merge(restaurants, on="restaurant_id", how="left")

final_df.head()
```

```python
final_df[final_df["membership"] == "Gold"] \
    .groupby("city")["total_amount"] \
    .sum() \
    .sort_values(ascending=False)
```

The final merged dataset is used as the single source of truth for analysis.
Insights such as revenue trends, user behavior, and membership impact
can be derived using aggregation and grouping techniques.
